# Problem 1

## List the case number, type of crime and community area for all crimes in community area number 18.

>
Hint 1
Use tables CHICAGO_CRIME_DATA and CENSUS_DATA.
>
Hint 2
Use an inner join.
>
Hint 3
The column PRIMARY_TYPE contains the crime type.

```python
%%sql 

select A.case_number,A.primary_type,A.community_area_number, B.community_area_name
from CHICAGO_CRIME_DATA as A
Inner Join CENSUS_DATA as B 
on B.community_area_number = A.community_area_number
AND A.community_area_number = '18'

```

# Problem 2

## List all crimes that took place at a school. Include case number, crime type and community name.
>
Hint 1
Use tables CHICAGO_CRIME_DATA and CENSUS_DATA.
>
Hint 2
Use a left join or a right join.
>
Hint 3
The column LOCATION_DESCRIPTION will help you find the crime location.

```python
%%sql 

select A.case_number,A.primary_type, B.community_area_name
from CHICAGO_CRIME_DATA as A 
Left Join CENSUS_DATA as B 
on A.community_area_number = B.community_area_number
where A.LOCATION_DESCRIPTION Like '%SCHOOL%'
```

# Problem 3

## For the communities of Oakland, Armour Square, Edgewater and CHICAGO list the associated community_area_numbers and the case_numbers.

>
Hint 1
Use tables CHICAGO_CRIME_DATA and CENSUS_DATA.
>
Hint 2
Use a full outer join.
>
Hint 3
Use COMMUNITY_AREA_NUMBER from CENSUS_DATA.

```python
%%sql 

select A.case_number,B.community_area_name,B.community_area_number
from CHICAGO_CRIME_DATA as A
Full Outer Join CENSUS_DATA as B 
on A.community_area_number = B.community_area_number
where B.community_area_name in ('Oakland', 'Armour Square', 'Edgewater', 'CHICAGO')

```
